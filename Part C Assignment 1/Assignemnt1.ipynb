Aim : Create a review scrapper for any ecommerce website to fetch real time comments, reviews, rating, comment tags, customer name using python

importing Required Libraries

```python
import requests #Use to send HTTP request to a specified URL
import bs4 #used for pulling data from HTML and XML file
```

Getting data Requests

```python
request1 = requests.get('https://www.flipkart.com/poco-c31-royal-blue-64-gb/p/itm19effae969b86')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
# getting the data in text format
```

Fetching reviews / Comments

```python
#Fetch Reviews / Comments
reviews = soup.find_all('div',{'class' : 'ZmyHeo'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

Fetching Ratings

```python
#Average Overall Rating
ratings = soup.find('div',{'class' : 'ipqd2A'}).get_text();
print(ratings)
```

```python
#Individual Rating
individual_ratings = soup.find_all('div',{'class' : 'XQDdHH Ga3i8K'})
for individual_rating in individual_ratings:
    print(individual_rating.get_text() + "\n\n")
```

Fetching Tags

```python
#Fetch Tags
tags = soup.find('span',{'class' : '_6EBuvT OHdLLG'}).get_text();
tags
```

Fetching Customer Names

```python
#Fetch Customer Name
customer_names = soup.find_all('p',{'class' : '_2NsDsF AwS1CA'})
for customer_name in customer_names:
    print(customer_name.get_text() + "\n\n")
```
